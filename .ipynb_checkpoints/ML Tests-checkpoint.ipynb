```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import tree
import glob
import os
import pandas as pd
import csv
import cv2 
import time

files_wave = "./maps/wave/*.csv"
dir_wave = "./maps/wave"

files_power = "./maps/power/*.csv"
dir_power = "./maps/power"

files_smsr = "./maps/smsr/*.csv"
dir_smsr = "./maps/smsr"


df = pd.read_csv("./labels.csv", usecols=["Id","Category"])

def plot3Maps(maps, filenames):
    plt.subplot(131), plt.imshow(maps[0], cmap="gray")
    plt.title(filenames[0]), plt.xticks([]), plt.yticks([])
    plt.subplot(132), plt.imshow(maps[1], cmap="gray")
    plt.title(filenames[1]), plt.xticks([]), plt.yticks([])
    plt.subplot(133), plt.imshow(maps[2], cmap="gray")
    plt.title(filenames[2]), plt.xticks([]), plt.yticks([])
    plt.show()
    
def dataToNormalizedImage(maps):
    maps = np.float32(maps)
    maps = maps-np.amin(maps)
    maps = (maps/np.amax(maps))*255
    maps = np.uint8(maps)
    return maps

def getMaps(dirc,files):
    train_maps = np.array([list(csv.reader(open(maps))) for maps in glob.glob(files)])
    filenames = os.listdir(dirc)
    return train_maps , filenames

def combineMaps(wave,power,smsr):
    numMaps = wave.shape[0]+power.shape[0]+smsr.shape[0]
    data = np.zeros((numMaps,wave.shape[1],wave.shape[2]))
    data[0:wave.shape[0]] = wave[0:wave.shape[0]]
    data[wave.shape[0]:wave.shape[0]+power.shape[0]] = power[0:power.shape[0]]
    data[wave.shape[0]+power.shape[0]:numMaps] = smsr[0:smsr.shape[0]]
    data = np.uint8(data)
    return data

wave_maps, wave_files  = getMaps(dir_wave,files_wave)
wavemaps = wave_maps.copy()
wave_maps = dataToNormalizedImage(wave_maps)

pow_maps, pow_files = getMaps(dir_power,files_power)
pow_maps = dataToNormalizedImage(pow_maps)

smsr_maps, smsr_files = getMaps(dir_smsr, files_smsr)
smsr_maps = dataToNormalizedImage(smsr_maps)

final = combineMaps(wave_maps,pow_maps,smsr_maps)

plot3Maps(wave_maps, wave_files)
plot3Maps(pow_maps, pow_files)
plot3Maps(smsr_maps, smsr_files)


##########################################################################
sift = cv2.xfeatures2d.SIFT_create()

# Instantiate bag-of-words
dictionarySize = 20 # number of features 
BOW = cv2.BOWKMeansTrainer(dictionarySize)
    
# Find keypoints and descriptors with SIFT for each training image
for i, img in enumerate(final):
    kp, dsc = sift.detectAndCompute(img, None)
    print(kp)
    print(dsc)
   # BOW.add(dsc)
    print(i)

start_time = time.time()
codebook = BOW.cluster()
print("Time required for clustering: ", time.time() - start_time)

```

```python
tenth_maps = (np.float32(moop[:, :56*21])*10).astype(int)



def mixMaps(maps, first, second, third):
    mixed = np.zeros(56*21)
    for i in range(56*21):
        check = np.random.rand()*100
        if((check>0) and (check<first)):
            mixed[i] = maps[0,i]
            
        elif((check>first) and (check<(first+second))):
            mixed[i] = maps[1,i]
            
        elif((check>(first+second))): 
            mixed[i] = maps[2,i]
    
    return mixed.astype(int)
        
    
tester = mixMaps(tenth_maps,33,33,33)  

mix_df = pd.DataFrame()
mix_df['dbr'] = x[:56*21,0]
mix_df['gain'] = x[:56*21,1]
mix_df['label'] = tester

features = mix_df[['dbr','gain']]
output = mix_df.label
mix_df.head()

mX_train, mX_test, my_train, my_test = train_test_split(features, output, test_size=0.1, random_state=1)
mix_df.head()

    
```
